```python
plt.figure(figsize = (10,5))
plt.title('Alive, dead divided by race')
sns.countplot(data = breast_cancer_data, x="Race", hue="Status") 


number_people_white = breast_cancer_data['Race'].value_counts()['White']
alive_white = breast_cancer_data.loc[breast_cancer_data['Status'] == 'Alive'][breast_cancer_data['Race'] == 'White']
alive_white_number = alive_white['Status'].value_counts()['Alive']
percent_white_alive = round(100 * (alive_white_number / number_people_white))

dead_white = breast_cancer_data.loc[breast_cancer_data['Status'] == 'Dead'][breast_cancer_data['Race'] == 'White']
dead_white_number = dead_white['Status'].value_counts()['Dead']
percent_white_dead = round( 100 * (dead_white_number / number_people_white))
number_people_black = breast_cancer_data['Race'].value_counts()['Black']
alive_black = breast_cancer_data.loc[breast_cancer_data['Status'] == 'Alive'][breast_cancer_data['Race'] == 'Black']
alive_black_number = alive_black['Status'].value_counts()['Alive']
percent_black_alive = round( 100 * (alive_black_number / number_people_black))

dead_black = breast_cancer_data.loc[breast_cancer_data['Status'] == 'Dead'][breast_cancer_data['Race'] == 'Black']
dead_black_number = dead_black['Status'].value_counts()['Dead']
percent_black_dead = round(100 * dead_black_number / number_people_black)

number_people_other = breast_cancer_data['Race'].value_counts()['Other']
alive_other = breast_cancer_data.loc[breast_cancer_data['Status'] == 'Alive'][breast_cancer_data['Race'] == 'Other']
alive_other_number = alive_other['Status'].value_counts()['Alive']
percent_other_alive = round( 100 * (alive_other_number / number_people_other))

dead_other = breast_cancer_data.loc[breast_cancer_data['Status'] == 'Dead'][breast_cancer_data['Race'] == 'Other']
dead_other_number = dead_other['Status'].value_counts()['Dead']
percent_other_dead = round(100 * dead_other_number / number_people_other)
print(f'Percentage Whites people which survived: {percent_white_alive}%; Dead : {percent_white_dead}%')
print(f'Percentage Blacks people which survived: {percent_black_alive}%; Dead : {percent_black_dead}%')
print(f'Percentage Other people which survived: {percent_other_alive}%; Dead : {percent_other_dead}%')
```
